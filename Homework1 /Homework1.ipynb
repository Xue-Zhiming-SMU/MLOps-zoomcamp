```python
import pyarrow.parquet as pq
```

```python
Jan = pq.read_table(source = 'yellow_tripdata_2023-01.parquet')
Feb = pq.read_table(source = 'yellow_tripdata_2023-02.parquet')

df_Jan = Jan.to_pandas()
df_Feb = Feb.to_pandas()
```

```python
df_Jan_Clean = df_Jan.copy()
df_Jan
```

```python
df_Jan_Clean['duration'] = (df_Jan['tpep_dropoff_datetime'] - df_Jan['tpep_pickup_datetime']).dt.total_seconds()/60
```

```python
import numpy as np
std = np.std(df_Jan_Clean['duration'])     
print(std)
```

```python
df_Jan_Clean = df_Jan_Clean.query('1 <= duration <= 60')
```

```python
# 1. Convert location IDs to strings
df_Jan_Clean[['PULocationID', 'DOLocationID']] = df_Jan_Clean[['PULocationID', 'DOLocationID']].astype(str)

# 2. Create list of dictionaries
dicts = df_Jan_Clean[['PULocationID', 'DOLocationID']].to_dict(orient='records')

# 3. Initialize and fit DictVectorizer
from sklearn.feature_extraction import DictVectorizer
dv = DictVectorizer()
X = dv.fit_transform(dicts)

# 4. Check dimensionality
print(f"Feature matrix shape: {X.shape}")  # (n_samples, n_features)
```

```python
y = df_Jan_Clean['duration'].values
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# 1. Initialize and train the model
lr = LinearRegression()
lr.fit(X, y)  # This is where the actual training happens

# 2. Make predictions on training data
y_pred = lr.predict(X)

# 3. Calculate RMSE
rmse = np.sqrt(mean_squared_error(y, y_pred))
print(f"Training RMSE: {rmse:.2f} minutes")
```

```python
# Process February data same as January data
df_Feb['duration'] = (df_Feb['tpep_dropoff_datetime'] - df_Feb['tpep_pickup_datetime']).dt.total_seconds()/60
df_Feb_Clean = df_Feb[(df_Feb['duration'] >= 1) & (df_Feb['duration'] <= 60)]

# Apply same DictVectorizer transformation
val_dicts = df_Feb_Clean[['PULocationID', 'DOLocationID']].astype(str).to_dict(orient='records')
X_val = dv.transform(val_dicts)  # Use transform() not fit_transform()
y_val = df_Feb_Clean['duration'].values
```

```python
# Make predictions and calculate RMSE
y_val_pred = lr.predict(X_val)
val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
print(f"Validation RMSE: {val_rmse:.2f} minutes")
```
